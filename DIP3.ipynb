```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow


I = cv2.imread('/content/hist5.bmp')


rmin = I.min()
rmax = I.max()  #max and min pixel values


S = (I - rmin) / (rmax - rmin) #contrast stretch formula
S = np.uint8(S * 255) #stretch it to 255 pixel value


print("Original Image")
cv2_imshow(np.uint8(I))

print("Contrast Stretched Image")
cv2_imshow(S)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

gray = cv2.imread("/content/Gray level slicing.bmp", cv2.IMREAD_GRAYSCALE)

A = 100
B = 180

sliced = np.zeros_like(gray)  #we are making the whole image black by making the pixel value of gray pixels 0
sliced[(gray >= A) & (gray <= B)] = 255

cv2_imshow(gray)
cv2_imshow(sliced)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

gray = cv2.imread("/content/Bit plane slicing.bmp", cv2.IMREAD_GRAYSCALE)

for bit in range(8):
    print("Bit plane:", bit)
    bp = (gray >> bit) & 1
    cv2_imshow(bp * 255)


cv2_imshow(bit)
```
