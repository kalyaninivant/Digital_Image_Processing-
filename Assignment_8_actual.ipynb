```python
# =========================================================
# 1. IMPORT LIBRARIES
# =========================================================
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt


# =========================================================
# 2. UPLOAD IMAGE
# =========================================================
uploaded = files.upload()

# Get uploaded file name
image_path = list(uploaded.keys())[0]

# Read image in grayscale
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)


# =========================================================
# 3. PREPROCESSING
# =========================================================
# Convert to float32 (IMPORTANT for DCT)
img = np.float32(img)

# Get size
h, w = img.shape

# Make dimensions divisible by 8
h = h - (h % 8)
w = w - (w % 8)
img = img[:h, :w]


# =========================================================
# 4. INITIALIZE OUTPUT ARRAYS
# =========================================================
dct_image = np.zeros((h, w))
reconstructed = np.zeros((h, w))


# =========================================================
# 5. BLOCK-WISE DCT & IDCT (8x8)
# =========================================================
for i in range(0, h, 8):
    for j in range(0, w, 8):

        # Extract 8x8 block
        block = img[i:i+8, j:j+8]

        # Apply DCT
        dct_block = cv2.dct(block)
        dct_image[i:i+8, j:j+8] = dct_block

        # Apply IDCT (Reconstruction)
        idct_block = cv2.idct(dct_block)
        reconstructed[i:i+8, j:j+8] = idct_block


# =========================================================
# 6. PREPARE DCT IMAGE FOR DISPLAY (LOG SCALE)
# =========================================================
dct_log = np.log(np.abs(dct_image) + 1)

# Normalize to 0–255
dct_norm = cv2.normalize(dct_log, None, 0, 255, cv2.NORM_MINMAX)
dct_norm = np.uint8(dct_norm)


# =========================================================
# 7. CONVERT RECONSTRUCTED IMAGE TO UINT8
# =========================================================
reconstructed = np.clip(reconstructed, 0, 255)
reconstructed = np.uint8(reconstructed)


# =========================================================
# 8. DISPLAY RESULTS
# =========================================================
plt.figure(figsize=(15,5))

# Original Image
plt.subplot(1,3,1)
plt.title("Original Image")
plt.imshow(img, cmap='gray')
plt.axis('off')

# DCT Image
plt.subplot(1,3,2)
plt.title("DCT Image (8x8 blocks)")
plt.imshow(dct_norm, cmap='gray')
plt.axis('off')

# Reconstructed Image
plt.subplot(1,3,3)
plt.title("Reconstructed Image (IDCT)")
plt.imshow(reconstructed, cmap='gray')
plt.axis('off')

plt.show()
```
