```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/hist1.bmp', 0)


hist = np.zeros(255)
for pixel in img.flatten():
    hist[pixel] += 1


pdf = hist / hist.sum()


cdf = np.zeros(255)
cdf[0] = pdf[0]
for i in range(1, 255):
    cdf[i] = cdf[i-1] + pdf[i]


L = 255
mapping = np.round((L - 1) * cdf).astype('uint8')


equ_img = mapping[img]


cv2_imshow(img)
cv2_imshow(equ_img)



plt.figure()
plt.plot(hist, label='Histogram')
plt.plot(cdf * hist.max(), label='CDF (scaled)')
plt.legend()
plt.title("Histogram and CDF")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import os
from google.colab.patches import cv2_imshow
save_folder = "/content/drive/MyDrive/DIP/Histogram"
os.makedirs(save_folder, exist_ok=True)

valid_ext = (".jpg", ".jpeg", ".png", ".bmp", ".gif")

for filename in os.listdir(folder_path):
    if filename.lower().endswith(valid_ext):

        img_path = os.path.join(folder_path, filename)
        img = cv2.imread(img_path, 0)

        hist = cv2.calcHist([img],[0],None,[256],[0,256])

        plt.figure()
        plt.plot(hist)
        plt.title(filename)

        save_path = os.path.join(save_folder, filename + "_hist.png")
        plt.savefig(save_path)
        plt.close()
```
